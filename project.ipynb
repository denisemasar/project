```python
# ==========================================================
# TRATAMENTO DE DADOS
# ==========================================================

# remover linhas com valores nulos
df = df.dropna()

# remover linhas duplicadas
df = df.drop_duplicates()

# verificar novamente se existem nulos
df.isnull().sum().head(10)
```

```python
# aqui eu removo colunas que possuem apenas um único valor
# colunas assim não ajudam na análise

df = df.loc[:, df.nunique() > 1]
```

```python
# aqui estou criando uma nova coluna
# ela representa o total de gols da partida

df['total_goals'] = df['number of goals team1'] + df['number of goals team2']
```

```python
# ==========================================================
# PERGUNTA 1:
# Quais os times marcaram mais gols (melhores ataques)? (cite os 10 primeiros)
# ==========================================================

# Somar gols quando o time aparece como team1
gols_team1 = df.groupby('team1')['number of goals team1'].sum()

# Somar gols quando o time aparece como team2
gols_team2 = df.groupby('team2')['number of goals team2'].sum()

# Somar os dois resultados
total_gols = gols_team1.add(gols_team2, fill_value=0)

resultado = total_gols.sort_values(ascending=False).head(10).reset_index()
resultado.columns = ['Time', 'Total de Gols']

display(resultado)
```

```python
# ==========================================================
# PERGUNTA 2:
# Quais os times marcaram sofreram mais gols (defesas mais vazadas)? (cite os 10 primeiros)
# ==========================================================

# Somar gols que o team1 tomou, somando o gols do outro lado (team2)
gols_team1 = df.groupby('team1')['number of goals team2'].sum()

# Somar gols que o team2 tomou, somando o gols do outro lado (team1)
gols_team2 = df.groupby('team2')['number of goals team1'].sum()

# Somar os dois resultados
total_gols = gols_team1.add(gols_team2, fill_value=0)

resultado = total_gols.sort_values(ascending=False).head(10).reset_index()
resultado.columns = ['Time', 'Total de Gols']

display(resultado)
```

```python
# ==========================================================
# PERGUNTA 3:
# Qual os jogos tiveram mais gols (cite os 10 primeiros)?
# ==========================================================

# cria uma coluna com o total de gols da partida
df['total_gols'] = df['number of goals team1'] + df['number of goals team2']

# ordena os jogos pelo total de gols (do maior para o menor)
top10_jogos_gols = df.sort_values(by='total_gols', ascending=False)

# mostra os 10 jogos com mais gols
resultado = top10_jogos_gols[['team1', 'team2', 'number of goals team1', 'number of goals team2', 'total_gols']].head(10)

display(resultado)
```

```python
# ==========================================================
# PERGUNTA 4: Qual jogo teve a maior diferença de gols (maior goleada)?
# ==========================================================

# Calcular a diferença de gols usando valor absoluto (função "abs" = |módulo| )
df['goal_difference'] = abs(df['number of goals team1'] - df['number of goals team2'])

# Encontrar o jogo com a maior diferença de gols (maior goleada)
max_diff_game = df.loc[df['goal_difference'].idxmax()]

# Mostrar o resultado
print("Jogo com maior diferença de gols:")
print(f"{max_diff_game['team1']} {max_diff_game['number of goals team1']} x {max_diff_game['number of goals team2']} {max_diff_game['team2']}")
print(f"Diferença: {max_diff_game['goal_difference']} gols")
```

```python
# ==========================================================
# PERGUNTA 5:
# Quais os times tiveram a MAIOR POSSE DE BOLA MÉDIA (cite os 10 primeiros)?
# ==========================================================

# Função para calcular a posse média considerando ambos os times
def average_possession(df):
    # --- Preparar posse de team1 ---
    df['possession team1'] = df['possession team1'].astype(str).str.replace('%','', regex=False)
    df['possession team1'] = pd.to_numeric(df['possession team1'], errors='coerce')
    
    # --- Preparar posse de team2 ---
    df['possession team2'] = df['possession team2'].astype(str).str.replace('%','', regex=False)
    df['possession team2'] = pd.to_numeric(df['possession team2'], errors='coerce')
    
    # Média da posse de cada time
    posse1 = df.groupby('team1')['possession team1'].mean()
    posse2 = df.groupby('team2')['possession team2'].mean()
    
    # Combinar as médias (soma das duas quando o time aparece em team1 ou team2)
    posse_total = posse1.add(posse2, fill_value=0)
    
    # Transformar em DataFrame limpo
    posse_df = posse_total.reset_index()
    posse_df.columns = ['Time', 'Maior Posse de Bola Média']
    
    # Retornar os 10 maiores
    return posse_df.sort_values('Maior Posse de Bola Média', ascending=False).head(10)

# Usar a função
average_possession(df)
```

```python
# ==========================================================
# PERGUNTA 6:
# Quais os times deram mais CHUTES A GOL (cite os 10 primeiros)?
# ==========================================================

def total_shots_on_target(df):
    # Somatório dos chutes quando o time é team1
    shots_team1 = df.groupby('team1')['on target attempts team1'].sum()
    # Somatório dos chutes quando o time é team2
    shots_team2 = df.groupby('team2')['on target attempts team2'].sum()
    # Somar os dois
    total_shots = shots_team1.add(shots_team2, fill_value=0)
    # Transformar em DataFrame limpo
    shots_df = total_shots.reset_index()
    shots_df.columns = ['Time', 'Total de Chutes a Gol']
    return shots_df

total_shots_on_target(df).sort_values('Total de Chutes a Gol', ascending=False).head(10)
```

```python
# ==========================================================
# PERGUNTA 7:
# Quais os times deram mais PASSES (cite os 10 primeiros)?
# ==========================================================
#  #soma total de passes realizados por cada time

def total_passes(df):
    # Somar passes quando o time é team1
    passes_team1 = df.groupby('team1')['passes team1'].sum()
    # Somar passes quando o time é team2
    passes_team2 = df.groupby('team2')['passes team2'].sum()
    # Somar os dois
    total_passes = passes_team1.add(passes_team2, fill_value=0)
    # Transformar em DataFrame limpo
    passes_df = total_passes.reset_index()
    passes_df.columns = ['Time', 'Total Passes']
    return passes_df

total_passes(df).sort_values('Total Passes', ascending=False).head(10)
```

```python
# PERGUNTA 8:
# Quais os jogos tiveram mais CARTÕES AMARELOS (cite os 10 primeiros)?
# ==========================================================

# criar coluna com total de cartões amarelos da partida
df['total_yellow'] = df['yellow cards team1'] + df['yellow cards team2']

# ordenar pelos jogos com mais cartões
top10_cartoes = df.sort_values(by='total_yellow', ascending=False)

# selecionar colunas relevantes e pegar os 10 primeiros
resultado = top10_cartoes[['team1', 'team2','yellow cards team1','yellow cards team2','total_yellow']].head(10)

# exibir formatado
display(resultado)
```

```python
# ==========================================================
# PERGUNTA 9:
# Quais os times sofreram mais FALTAS (cite os 10 primeiros)?
# ==========================================================
# soma total de faltas sofridas por cada time

def fouls_suffered(df):
    # Faltas sofridas quando o time é team1
    fouls_team1 = df.groupby('team1')['fouls against team1'].sum()
    # Faltas sofridas quando o time é team2
    fouls_team2 = df.groupby('team2')['fouls against team2'].sum()
    # Somar os dois
    total_fouls = fouls_team1.add(fouls_team2, fill_value=0)
    # Transformar em DataFrame limpo
    fouls_df = total_fouls.reset_index()
    fouls_df.columns = ['Time', 'Faltas sofridas']
    return fouls_df

fouls_suffered(df).sort_values('Faltas sofridas', ascending=False).head(10)
```

```python
# ==========================================================
# PERGUNTA 10:
# Quais os times tiveram mais ESCANTEIOS (cite os 10 primeiros)?
# ==========================================================
def total_corners(df):
    # Escanteios quando o time é team1
    corners_team1 = df.groupby('team1')['corners team1'].sum()
    # Escanteios quando o time é team2
    corners_team2 = df.groupby('team2')['corners team2'].sum()
    # Somar os dois
    total_corners = corners_team1.add(corners_team2, fill_value=0)
    # Transformar em DataFrame limpo
    corners_df = total_corners.reset_index()
    corners_df.columns = ['Time', 'Total Escanteios']
    return corners_df

total_corners(df).sort_values('Total Escanteios', ascending=False).head(10)
```

```python
# ==========================================================
# PERGUNTA 11:
# Quais os times tiveram mais IMPEDIMENTOS (cite os 10 primeiros)?
# ==========================================================
# soma total de impedimentos por time

def total_offsides(df):
    # Impedimentos quando o time é team1
    offsides_team1 = df.groupby('team1')['offsides team1'].sum()
    # Impedimentos quando o time é team2
    offsides_team2 = df.groupby('team2')['offsides team2'].sum()
    # Somar os dois
    total_offsides = offsides_team1.add(offsides_team2, fill_value=0)
    # Transformar em DataFrame limpo
    offsides_df = total_offsides.reset_index()
    offsides_df.columns = ['Time', 'Total de Impedimentos']
    return offsides_df

total_offsides(df).sort_values('Total de Impedimentos', ascending=False).head(10)
```

```python
# ==========================================================
# PERGUNTA 12:
# Quais os times apresentaram maior PRESSÃO DEFENSIVA (cite os 10 primeiros)?
# ==========================================================

# soma total de pressões defensivas aplicadas por time

def total_defensive_pressure(df):
    # Pressão defensiva quando o time é team1
    pressure_team1 = df.groupby('team1')['defensive pressures applied team1'].sum()
    # Pressão defensiva quando o time é team2
    pressure_team2 = df.groupby('team2')['defensive pressures applied team2'].sum()
    # Somar os dois
    total_pressure = pressure_team1.add(pressure_team2, fill_value=0)
    # Transformar em DataFrame limpo
    pressure_df = total_pressure.reset_index()
    pressure_df.columns = ['Team', 'Total Defensive Pressure']
    return pressure_df

total_defensive_pressure(df).sort_values('Total Defensive Pressure', ascending=False).head(10)
```
