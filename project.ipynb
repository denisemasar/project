```python
# ==========================================================
# IMPORTAÇÃO DAS BIBLIOTECAS
# ==========================================================

# pandas = biblioteca usada para manipular dados
import pandas as pd
```

```python
# ==========================================================
# CARREGAR O DATASET
# ==========================================================

# leitura do arquivo csv
# se o arquivo estiver na mesma pasta do notebook basta usar o nome

df = pd.read_csv("fifa_world_cup.csv")

# visualizar as primeiras linhas para entender a estrutura
df.head(10)
```

```python
# shape mostra o tamanho da base de dados
# primeiro número = quantidade de linhas
# segundo número = quantidade de colunas
df.shape
```

```python
# lista todas as colunas do dataset
# isso ajuda a entender quais variáveis estão disponíveis para análise

df.columns
```

```python
# aqui eu verifico o tipo de cada coluna
# isso é importante para saber se os dados são números ou texto

df.dtypes
```

```python
# o info mostra um resumo geral da tabela
# aqui dá para ver quantidade de valores e se existem dados faltando

df.info()
```

```python
# calcula estatísticas das colunas numéricas
# inclui média, mínimo, máximo e desvio padrão

df.describe()
```

```python
# aqui estou verificando se existem dados faltando
# o isnull identifica valores vazios
# o sum conta quantos existem em cada coluna

df.isnull().sum()
```

```python
# ==========================================================
# VERIFICAR VALORES DISTINTOS
# ==========================================================

# quantos valores diferentes existem em algumas colunas
df.nunique().head(10)
```

```python
# ==========================================================
# TRATAMENTO DE DADOS
# ==========================================================

# remover linhas com valores nulos
df = df.dropna()

# remover linhas duplicadas
df = df.drop_duplicates()

# verificar novamente se existem nulos
df.isnull().sum().head(10)
```

```python
# aqui eu removo colunas que possuem apenas um único valor
# colunas assim não ajudam na análise

df = df.loc[:, df.nunique() > 1]
```

```python
# aqui estou criando uma nova coluna
# ela representa o total de gols da partida

df['total_goals'] = df['number of goals team1'] + df['number of goals team2']
```

```python
# ==========================================================
# PERGUNTA 1
# Quais os times marcaram mais gols? (cite os 10 primeiros)
# ==========================================================
df.groupby('team1')['number of goals team1'].sum().sort_values(ascending=False).head(10)

```

```python
# ==========================================================
# PERGUNTA 2
# Qual o jogo teve mais gols?
# ==========================================================
# ordeno os jogos pelo total de gols
# head(1) mostra apenas o jogo com maior valor

df.sort_values('total_goals', ascending=False).head(1)
```

```python
# ==========================================================
# PERGUNTA 3:
# Quais os times tiveram a MAIOR POSSE DE BOLA MÉDIA (cite os 10 primeiros)?
# ==========================================================

# primeiro vou garantir que a coluna esteja em formato texto
# isso evita erro caso alguns valores sejam números
df['possession team1'] = df['possession team1'].astype(str)

# agora removo o símbolo %
# regex=False evita problemas de interpretação
df['possession team1'] = df['possession team1'].str.replace('%', '', regex=False)

# converto novamente para número
# errors='coerce' transforma valores problemáticos em NaN
df['possession team1'] = pd.to_numeric(df['possession team1'], errors='coerce')

# agora sim posso calcular a média de posse por time
posse_media = (
    df.groupby('team1')['possession team1']
    .mean()
    .sort_values(ascending=False)
)

# mostrar resultado
posse_media.head(10)
```

```python
# ==========================================================
# PERGUNTA 4:
# Quais os times deram mais CHUTES A GOL (cite os 10 primeiros)?
# ==========================================================

df.groupby('team1')['on target attempts team1'].sum().sort_values(ascending=False).head(10)
```

```python
# ==========================================================
# PERGUNTA 5:
# Quais os times deram mais PASSES (cite os 10 primeiros)?
# ==========================================================
#  #soma total de passes realizados por cada time

df.groupby('team1')['passes team1'].sum().sort_values(ascending=False).head(10)
```

```python
# ==========================================================
# PERGUNTA 6:
# Quais os jogos tiveram mais CARTÕES AMARELOS (cite os 10 primeiros)?
# ==========================================================
# primeiro crio uma coluna com o total de cartões da partida

df['total_yellow'] = df['yellow cards team1'] + df['yellow cards team2']

# depois ordeno para encontrar o jogo com mais cartões

df.sort_values('total_yellow', ascending=False).head(10)
```

```python
# ==========================================================
# PERGUNTA 7:
# Quais os times sofreram mais FALTAS (cite os 10 primeiros)?
# ==========================================================
# soma total de faltas sofridas por cada time

df.groupby('team1')['fouls against team1'].sum().sort_values(ascending=False).head(10)
```

```python
# ==========================================================
# PERGUNTA 8:
# Quais os times tiveram mais ESCANTEIOS (cite os 10 primeiros)?
# ==========================================================
# soma total de escanteios por time

df.groupby('team1')['corners team1'].sum().sort_values(ascending=False).head(10)
```

```python
# ==========================================================
# PERGUNTA 9:
# Quais os times tiveram mais IMPEDIMENTOS (cite os 10 primeiros)?
# ==========================================================
# soma total de impedimentos por time

df.groupby('team1')['offsides team1'].sum().sort_values(ascending=False).head(10)
```

```python
# ==========================================================
# PERGUNTA 10:
# Quais os times apresentaram maior PRESSÃO DEFENSIVA (cite os 10 primeiros)?
# ==========================================================

# soma total de pressões defensivas aplicadas por time

df.groupby('team1')['defensive pressures applied team1'].sum().sort_values(ascending=False).head(10)
```
